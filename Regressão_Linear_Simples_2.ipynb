# Exercício de Regressão Linear Simples

## Importando as bibliotecas

```python
import pandas as pd
import numpy as np
import seaborn as sns
```


## Importando a base de dados



```python
alunos = pd.read_excel("/content/Cópia de aprovacao_alunos.xlsx")
alunos.head()
```

## Criando um gráfico com a reta de regressão

```python
ax = sns.lmplot(x="ano_2015", y="ano_2016", data=alunos)
ax.fig.set_size_inches(12, 6)
ax.fig.suptitle("Reta de Regressão - 2015 x 2016", fontsize=16, y=1.02)
ax.set_xlabels("2015", fontsize=14)
ax.set_ylabels("2016", fontsize=14)
ax;
```

## Identificando a correlação entre as variáveis

```python
alunos.corr()
```

## Printando o resultado OLS Regression para entender o R^2, Intercepto e o parâmetro da variável explicativa

```python
import statsmodels.api as sm
Y = alunos.ano_2016
X = sm.add_constant(alunos.ano_2015)

resultado_regressao = sm.OLS(Y, X).fit()

print(resultado_regressao.summary())
```
