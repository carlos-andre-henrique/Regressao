# REGRESSÃO LINEAR MÚLTIPLA

## Importando as bibliotecas

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

## Importando a base de dados

```python
limite = pd.read_excel("/content/Cópia de Limite_Credito_Dummy.xlsx")
limite.head()
```

## Transformando as variáveis categóricas em variáveis indicadoras

```python
limite = pd.get_dummies(data=limite, columns=['Escolaridade', 'Gênero', 'Região'], prefix = ['Escolaridade', 'Região', 'Gênero'])
limite.head()
```

## Verificando a correlação entre as variáveis

```python
limite.corr()
```

## Criando e analisando os resultados da OLS

```python
import statsmodels.formula.api as smf

function = 'LimitedoChequeEspecial~Idade+RendimentoTotal+Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio+Escolaridade_Superior_Pos+Região_RJ+Região_SP+Gênero_F+Gênero_M'
model = smf.ols(formula=function, data=limite).fit()
print(model.summary())
```

## Removendo a variável Idade que possui o maior P-value

```python
function = 'LimitedoChequeEspecial~RendimentoTotal+Salario+LimitedeCreditoImediato+Escolaridade_Fundamental_Medio+Escolaridade_Superior_Pos+Região_RJ+Região_SP+Gênero_F+Gênero_M'
model = smf.ols(formula=function, data=limite).fit()
print(model.summary())
```

```python
function = 'LimitedoChequeEspecial~RendimentoTotal+Salario+LimitedeCreditoImediato+Escolaridade_Superior_Pos+Região_RJ+Região_SP+Gênero_F+Gênero_M'
model = smf.ols(formula=function, data=limite).fit()
print(model.summary())
```
